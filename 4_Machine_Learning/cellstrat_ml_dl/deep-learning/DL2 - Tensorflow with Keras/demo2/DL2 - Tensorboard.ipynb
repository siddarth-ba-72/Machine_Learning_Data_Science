```python
import tensorflow as tf
tf.compat.v1.disable_eager_execution()
import numpy as np
```

```python
x = tf.compat.v1.placeholder('float',name='X')
y=  tf.compat.v1.placeholder('float',name='y')
result = tf.add(x,y)
```

```python
with tf.compat.v1.Session() as sess:
    sess.run(tf.compat.v1.global_variables_initializer())
    for i in range(100):
        var1=  np.random.rand()
        var2=  np.random.rand()
        print(var1,var2)
        #tf.summary.scalar('result',sess.run(addition, feed_dict={x:var1,y:var2}))
        tf.compat.v1.summary.scalar('result',sess.run(result, feed_dict={x:var1,y:var2}))
        writer = tf.compat.v1.summary.FileWriter('Graphs',sess.graph)
```
