# TensorBoard

```python
import tensorflow as tf
import numpy as np
```

```python
x = tf.placeholder('float',name='X')
y=  tf.placeholder('float',name='y')
result = tf.add(x,y)
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for i in range(100):
        var1=  np.random.rand()
        var2=  np.random.rand()
        print(var1,var2)
        #tf.summary.scalar('result',sess.run(addition, feed_dict={x:var1,y:var2}))               
        tf.summary.scalar('result',sess.run(result, feed_dict={x:var1,y:var2}))               
        writer = tf.summary.FileWriter('Graphs',sess.graph)
```

### After the program is complete

In the command line prompt where the code is, type:    tensorboard --logdir Graphs
